```python
from pandas.io.formats.format import DataFrameFormatter
DataFrameFormatter=df[['eventid','iyear','imonth','iday','extended','country','country_txt','region','targsubtype1','attacktype1',
           'latitude','longitude','specificity','INT_IDEO','INT_MISC','INT_ANY']]
```

```python
df=df.rename(columns={'iyear':'Year','imonth':'Month', 'iday':'Day','extended':'Extended','country':'Country','country_txt':'Country in text','region':'Region','targsubtype1':'Targsub',
                          'attacktype1':'Attacks','latitude':'Lat','longitude':'Lon','specificity':'Specific','INT_IDEO':'Intideo',
                                 'INT_MISC':'Intmisc', 'INT_ANY':'Intany'})
```

```python
df['Intideo']=df['Intideo'].fillna('Unknown')
```

```python
df['Lat']=df['Lat'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(20,20))
heading=['Attacks In Day','Attacks In Month']
columns=['Day','Month']
for i,col in enumerate(columns):
  plt.subplot(2,1,i+1)
  plt.title(heading[i])
  sns.kdeplot(x=df[col])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,20))
heading=['Attacks In Day','Attacks In Month']
columns=['Day','Month']
for i,col in enumerate(columns):
  plt.subplot(2,1,i+1)
  plt.title(heading[i])
  sns.stripplot(x=df[col])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,20))
heading=['Attacks In Day','Attacks In Month']
columns=['Day','Month']
for i,col in enumerate(columns):
  plt.subplot(2,1,i+1)
  plt.title(heading[i],fontsize=20)
  plt.xlabel(col,fontsize=20)
  sns.swarmplot(x=df[col])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,20))
heading=['Attacks In Region','Attacks In Country']
columns=['Region','Country in text']
for i,col in enumerate(columns):
  plt.subplot(2,1,i+1)
  plt.title(heading[i],fontsize=20)
  plt.xticks(rotation=90)
  plt.xlabel(col,fontsize=20)
  sns.countplot(x=df[col])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.xticks(rotation=60)
plt.yticks(rotation=60)
plt.title('Attacks')
sns.violinplot(x=df['Attacks'],y=df['Day'])
```

```python
pd.crosstab(df.Year,df.Region).plot(kind='area',figsize=(15,8))
plt.title("ATTACKS IN EACH REGION")
plt.xlabel('Year')
plt.ylabel('Attacks')
plt.show()
```
