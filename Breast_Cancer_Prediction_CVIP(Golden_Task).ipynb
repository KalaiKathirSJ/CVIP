```python
for col in data.columns:
  a=data[col].quantile(0.25)
  b=data[col].quantile(0.75)
  iqr=b-a
  upper_limit=b+(1.5*iqr)
  lower_limit=a-(1.5*iqr)
  data[col] = np.where(data[col] < lower_limit, lower_limit,
  np.where(data[col] > upper_limit, upper_limit,data[col]))
```

```python
x=data.drop(['diagnosis'],axis=1)
y=data.diagnosis
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
standard_scaler=StandardScaler()
x_train=standard_scaler.fit_transform(x_train)
x_test=standard_scaler.transform(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
```

```python
algorithms = ['DecisionTreeClassifier','LogisticRegression','KNeighborsClassifier','RandomForestClassifier','SVC','AdaBoostClassifier']
value = [ DecisionTreeClassifier(criterion="entropy",random_state=42),LogisticRegression(), KNeighborsClassifier(n_neighbors = 9, weights ='uniform'),RandomForestClassifier(n_estimators=300), SVC(kernel="rbf",random_state=15), AdaBoostClassifier(n_estimators=100)]
models = dict(zip(algorithms,value))
models
```

```python
predicted =[]
for name,algo in models.items():
    model=algo
    model.fit(x_train,y_train)
    predict = model.predict(x_test)
    accuracy = accuracy_score(y_test, predict)*100
    predicted.append(accuracy)
    print(name,accuracy)
```

```python
plt.figure(figsize = (10,10))
sns.barplot(x = predicted, y = algorithms, palette='magma')
plt.title("ACCURACY VALUES", fontsize=20)
```
